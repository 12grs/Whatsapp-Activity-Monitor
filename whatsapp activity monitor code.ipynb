```python
import pandas as pd
from datetime import datetime
import openpyxl as op
waitforonline=WebDriverWait(driver,60)
chatwindow=driver.find_element_by_id('main')

intervalStart=0
isOnline=False
startTimes=[]
intervals=[]

scrutiny_time=60*60*4

while(scrutiny_time):
    
    nameNonline=chatwindow.find_element_by_class_name('_33QME').find_elements_by_tag_name('div')
    
    if(len(nameNonline)==3):
        
        if(not isOnline):
            
            intervalStart=time.perf_counter()
            t = time.localtime()
            startTimeOnline = time.strftime("%H:%M:%S", t)
            print(" is online at "+str(startTimeOnline))
        isOnline=True
    else:
        if(isOnline):
            isOnline=False

            print("Time:"+str(startTimeOnline)+" Time Interval:"+str(time.perf_counter()-intervalStart))
            startTimes.append(str(startTimeOnline))
            intervals.append(str(time.perf_counter()-intervalStart))
    
    time.sleep(1)
    scrutiny_time-=1
for i in range(len(startTimes)):
    print("Time:"+str(startTimes[i])+" Time Interval:"+str(intervals[i]))


table_header=["At time","For timeinterval"]


data_table=[]


for i in range(len(startTimes)):
    one_row=[startTimes[i],intervals[i]]
    data_table.append(one_row)
# print(data_table)

df=pd.DataFrame(data_table,columns=table_header) 
todaydatetime=str(datetime.now()).split()
todaydate=todaydatetime[0].split('-')
todaydate=todaydate[2]+"_"+todaydate[1]+"_"+todaydate[0]
sheetname=NAME+"_"+todaydate

with pd.ExcelWriter("WhatsappTrackerRecord.xlsx",engine="openpyxl",mode='a') as writer:
    df.to_excel(writer,sheet_name=sheetname)
df 
```
